```python
def detect_brand_from_model(car_name, existing_brand):
    if existing_brand != "Unknown":
        return existing_brand
    
    for keyword, brand in model_keyword_brand_map.items():
        if re.search(rf"\b{keyword}\b", car_name, flags=re.IGNORECASE):
            return brand
    return "Unknown"

df["Brand"] = df.apply(lambda row: detect_brand_from_model(row["Car Name"], row["Brand"]), axis=1)
```

```python
def extract_brand_from_name(car_name):
    # Remove the year (starts with 4 digits)
    name_no_year = re.sub(r"^\d{4}\s+", "", car_name)

    # Get first 1-2 words
    parts = name_no_year.split()
    if len(parts) >= 2:
        brand_guess = parts[0]
        
        # Handle double-word brand names
        if parts[0] in ["Hindustan", "Rolls", "Aston", "Land"]:
            brand_guess = f"{parts[0]} {parts[1]}"
        
        return brand_guess
    return "Unknown"
```
