```python
# Function to extract brand name
def extract_brand(car_name):
    parts = car_name.split()
    if len(parts) >= 2:
        return parts[1]  # Assuming format is: Year Brand Model...
    return None

# Apply the function to the 'car' column
df['brand'] = df['Car Name'].apply(extract_brand)
```

```python
def extract_brand_model(car_name):
    parts = car_name.split()
    if len(parts) >= 3:
        year = parts[0]
        brand = parts[1]
        
        # Check if brand is two words (like "Mercedes-Benz" or "Land Rover")
        if '-' in parts[1] or parts[1].lower() in ['land', 'rolls', 'alfa']: 
            brand = f"{parts[1]} {parts[2]}"
            model = ' '.join(parts[3:])
        else:
            model = ' '.join(parts[2:])
        return pd.Series([brand, model])
    return pd.Series([None, None])

# Apply and create new columns
df[['brand', 'model']] = df['Car Name'].apply(extract_brand_model)
```

```python
# Features to detect
import re
drive_features = {
    'manual_override': 'Manual Override',
    'paddle_shift': 'Paddle Shift',
    'sport_mode': 'Sport Mode'
}

# Extractor function
def extract_transmission_details(row):
    text = str(row)

    # Transmission Type (e.g., 'Automatic (DCT)', 'Manual', etc.)
    trans_match = re.match(r'^([^,-]+?)(?:\s*-\s*|\s*,)', text)
    transmission_type = trans_match.group(1).strip() if trans_match else None

    # Number of Gears
    gears_match = re.search(r'(\d+)\s*Gears', text)
    gears = gears_match.group(1) if gears_match else None

    # Extract features as boolean
    features = {key: feature in text for key, feature in drive_features.items()}

    return pd.Series([
        transmission_type,
        gears,
        features['manual_override'],
        features['paddle_shift'],
        features['sport_mode']
    ])

# Apply to DataFrame
df[['transmission_type', 'gears', 'manual_override', 'paddle_shift', 'sport_mode']] = df['Transmission'].apply(extract_transmission_details)
```
